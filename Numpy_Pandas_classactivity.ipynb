```python
#Second que
matrix1 = np.full((3,3),5) #to create a 3x3 matrix only with 5
array2 = np.arange(1,10)#creating array from 1 to 9
matrix2 = array2.reshape(3,3) #reshaping it into another matrix.
sum_result = np.add(matrix1, matrix2)
print('the sum is ')
print(sum_result)
prod_result = np.dot(matrix1,matrix2)
print('the product is')
print(prod_result)
sum_transpose = np.transpose(sum_result)
product_transpose = np.transpose(prod_result)
print('the transpose of sum is')
print(sum_transpose)
print('the transpose of product is')
print(product_transpose)
```

```python
arrr = np.random.rand(5*4)
arr_matrix = arrr.reshape(5,4)
print(arr_matrix)
row_sums = np.sum(arr_matrix, axis=1)
max_row_index = np.argmax(row_sums)

print("the row with maximum sum of values is", max_row_index)

print("the normalised matrix is")

for row in arr_matrix:
    for element in row:
        arr_norm = (element - np.min(arr_matrix,axis=1))/(np.max(arr_matrix,axis=1)-np.min(arr_matrix,axis=1))

    print(arr_norm)
```

```python
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
np.unique(arr)
values, counts = np.unique(arr, return_counts=True)
unique_values = values[counts == 1]
print("the unique values are:", unique_values)
duplicate_values = values[counts != 1]
print("the duplcate values are:", duplicate_values)

```

```python
main = np.arange(1,37)
main_matrix = main.reshape(6,6)
print("the central 4x4 matrix is")
central_matrix = main_matrix[1:5,1:5]
print(central_matrix)

print("the even elements are:")

for row in central_matrix:
    for element in row:
      if element%2 == 0:
        print(element)
      else:
        pass
```

```python
first_array = np.arange(0,12)
first_matrix = first_array.reshape(3,4)
reshaped = first_matrix.reshape(2,6)
print("the 2x2 array is\n",reshaped)

flat_reshaped = reshaped.flatten()
print("the flattened version is:\n",flat_reshaped)

second_matrix = flat_reshaped.reshape(4,3)
print("the 4x3 matrix is:\n", second_matrix)

```

```python
a = np.array([[1, 2], [3, 4]])
new_row = np.array([5, 6])
row_added_array = np.vstack((a,new_row))
print("the row added array is:\n",row_added_array)
new_column = np.array([7,8,9])
column_added_array = np.column_stack((row_added_array, new_column))
print("the column added array is:\n", column_added_array)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
joint_array = np.concatenate((a,b,c))
print("the joint array is:\n", joint_array)
reshaped_matrix = joint_array.reshape(3,3)
print("the 3x3 reshaped matrix is:\n", reshaped_matrix)
indices=np.argwhere(reshaped_matrix)
print("the diagonal elements are:\n")
for row,col in indices:
  element = reshaped_matrix[row,col]
  if row==col:
    print(element)

```

```python
arr_int = np.arange(1,21)
indices = np.argwhere(arr_int)
count=0
for i in indices:
  element = arr_int[i[0]]
  if element%3==0 or element%5==0:
    count=count+1
    print(element)
print("the count is:\n",count)
```
