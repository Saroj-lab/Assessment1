```python
data.drop_duplicates(keep='first')
```

```python
data = data.rename(columns={'Title': 'Movie_Title'})
data
```

```python
data['Genre'].str.split(",").explode().unique()
```

```python
data['Genres_per_movie'] = data['Genre'].str.split(',').apply(len)

data[['Movie_Title', 'Genres_per_movie']]
```

```python

data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].median(), inplace=True)

data['Metascore'].fillna(data['Metascore'].mean(), inplace=True)

data.isnull().sum()
```

```python
top_five_movies = data.nlargest(5, 'Rating')
top_five_movies
```

```python
data.loc[[data['Revenue (Millions)'].idxmin(), data['Revenue (Millions)'].idxmax()]]
```

```python
data.groupby('Director')['Rating'].mean()
```

```python
data.groupby('Year')['Movie_Title'].count()
```

```python
data[(data['Rating']>8) & (data['Revenue (Millions)'] > 100)][['Movie_Title']]
```

```python


data.groupby('Director').agg({
    'Rating': 'mean',
    'Revenue (Millions)': 'sum'
})
```

```python
data.groupby('Year').agg({
    'Movie_Title': 'count',
    'Rating': 'mean'
})
```

```python
data['Profitability'] = data['Revenue (Millions)'] / data['Runtime (Minutes)']
data['Profitability']
```

```python
data.nlargest(10, 'Profitability')[['Movie_Title', 'Profitability']]
```

```python
conditions = [
    (data['Rating'] < 6),
    (data['Rating'] >= 6) & (data['Rating'] < 7.5),
    (data['Rating'] >= 7.5)
]

choices = ['Poor', 'Average', 'Excellent']

data['Rating_Category'] = np.select(conditions, choices, default='Unknown')

data[['Movie_Title', 'Rating', 'Rating_Category']].head()
```

```python
def calculate_movie_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

data['Movie_Score'] = data.apply(calculate_movie_score, axis=1)

data[['Movie_Title', 'Rating', 'Revenue (Millions)', 'Movie_Score']].head()
```
